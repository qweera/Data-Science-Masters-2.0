```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

Ans.3 Error related to numeric calculations during the program execution are defined under ArithmeticError class. The ArithmeticError class contains the following subclasses under it:
FloatingPointError, OverflowError, ZeroDivisionError, DecimalException

```python
#Example for FloatingPoint Error
import sys
import math
#import fpectl 
try:
    print (math.exp(1000.0))
except Exception as e:
    print(e)
```

```python
#Example for ZeroDivision error
try:
    name=int(input("Enter a number"))
    10/name    
except ZeroDivisionError as e:
    print(e)
    
```

```python
#Example for Overflow Error
#When an arithmetic operation exceeds the limits of the variable datatype, an OverflowError is thrown.
i=1
try:
    f = 5.0**i
    for i in range(100):
        print('{},{}'.format(i, f))
        f = f ** 2
except OverflowError as e:
    print('Overflowed after {},{}'.format(f, e))
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

LookupError Exception is the base class for errors that need to be raised when something canâ€™t be found. And it can be raised when a key or index used on a mapping or sequence is invalid/notfound. The subclasses under LookupError( class are IndexError, KeyError, CodecRegistryError.

```python
# KeyError Example
my_dict = {'abc':23, 'bcd':['sss','ddd','rrr'],34:'sparrow'}
my_dict['abc']      #when a key is present
try:
    my_dict['hhh']    #when a key is not found
except KeyError as e:
    print(e)   #this will prints the missing key 
```

```python
#IndexError Example
my_fruits = ['apple', 'coconut', 'mango', 'pineapple', 'strawberry']
my_fruits[3]    #when given index is valid
try:
    my_fruits[5]    #when given index is out of range
except IndexError as e:
    print(e)
```

```python
Q6. List down some best practices for exception handling in python.
```

##### Ans.6 Some best practices for exception handling in python:
##### i) Always use generic exception and print a valid message, which will help others to understand and debug the code when required
##### ii) Always try to use logging as it will note down the errors and all activity permannenty in a file rather than storing it in the console using print statements 
##### iii) Always try to avoid writing multiple unneccessay exception handling statement for one exception, only write the exception which seems closed to you as it will increase the time for processing all the lines 
##### iv) Prepare a proper documentation, avoid using anything that will create problem for future developer who will try to modify. So proper documentation of entire code, classes, files, objects and methods that we'll use is very much required
##### v) Cleanup all the resources once their work is completed, as some resources will be allocated in the main memory
