### Q1. Explain GET and POST methods.

Ans1. GET  method is used to retrieve data from the server. e.g Google's search engine uses GET method, there is a search button to search a query. ### Q2. Why is request used in Flask?

Ans2. In Flask, the `request` object is used to access incoming HTTP request data sent by a client to a Flask web application. It allows you to access information such as form data, URL parameters, headers, and more. This object is essential for handling and processing user input, enabling your Flask application to respond to different client requests effectively.

### Q3. Why is redirect() used in Flask?

Ans3. In Flask, the `redirect()` function is used to send an HTTP response to the client, instructing it to navigate to a different URL. This is commonly used after processing a form submission or when you want to direct users to another part of your website. It helps in keeping your application organized and user-friendly by guiding users to relevant pages or actions.

### Q4. What are templates in Flask? Why is the render_template() function used?

Ans4. **Templates**: HTML files with placeholders for dynamic content. Templates in Flask are files containing HTML code with placeholders for dynamic data. They allow you to separate the presentation (HTML) from the logic (Python code) in your web application. 

**render_template()**: A Flask function that takes a template file as input and renders it as an HTTP response with dynamic data. It helps create dynamic web pages by filling in placeholders with data from your Python code.

Using `templates` and `render_template()` in Flask makes your code more organized, maintainable, and follows the separation of concerns principle by separating the HTML presentation from the Python logic.
