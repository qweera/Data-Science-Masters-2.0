#### Q3. What is App routing in Flask? Why do we use app routes?

Ans3. App routing in Flask is a process of mapping URLs to the associated Python functions intended to perform some task in a Flask application.

We use app routes in Flask for:

URL Handling: App routes allow you to specify how different URLs are handled by your Flask application. Each route is associated with a specific view function, which is responsible for processing the request and generating a response.

Modular Code: Using app routes helps you organize your code in a modular way. You can define separate view functions for different parts of your application and associate them with specific routes. This makes your code easier to manage and maintain.

Dynamic Routing: Flask supports dynamic routing, which means you can define routes with parameters (e.g., /user/<username>) to handle variable parts of URLs. These parameters can then be extracted and used within the view function to provide dynamic content.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans5. Flask url_for is defined as a function that enables developers to build and generate URLs on a Flask application.

```python

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/profile/<username>')
def profile(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    app.run(debug=True)
```

```python
from flask import Flask, url_for

app = Flask(__name__)

# Let's assume this is a different Python script

with app.test_request_context():
    # Generating URLs using url_for
    homepage_url = url_for('index')
    about_url = url_for('about')
    profile_url = url_for('profile', username='john_doe')

print("Homepage URL:", homepage_url)
print("About URL:", about_url)
print("Profile URL:", profile_url)
```
