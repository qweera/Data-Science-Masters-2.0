Ans2. Scatter plots are used to explore and visualize the relationship or distribution of data points between these two/three variables using 2D/3D graphs. Each dot represents a single data point, with its position on the graph determined by the values of two variables/three variables: one variable on the x-axis, another on the y-axis and another in z-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y, c='#E86917',marker='X')
plt.title("This is my X vs Y graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

Ans3. The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure or canvas. It allows you to arrange and organize multiple smaller plots in a grid-like fashion, making it easier to visualize and compare different aspects of your data in one coherent figure

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(8,7))
plt.subplot(2, 2, 1)  # Top-left subplot
plt.plot(x1, y1)
plt.title('Subplot 1')
plt.grid()

plt.subplot(2, 2, 2)  # Top-right subplot
plt.plot(x2, y2)
plt.title('Subplot 2')
plt.grid()

plt.subplot(2, 2, 3)  # Bottom-left subplot
plt.plot(x3, y3)
plt.title('Subplot 3')
plt.grid()

plt.subplot(2, 2, 4)  # Bottom-right subplot
plt.plot(x4, y4)
plt.title('Subplot 4')
plt.grid()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans4. A bar plot is a data visualization technique that uses rectangular bars to represent and compare the values of different categories or groups of data. Each bar's length or height is proportional to the value it represents. Using barplots we're able to plot categorical and numerical data. Bar plots are used to:

1. **Compare Data:** They are primarily used to compare values among different categories or groups, making it easy to identify variations, trends, or differences in data.

2. **Categorical Data:** Bar plots are particularly useful for visualizing categorical or discrete data where each category represents a distinct group or label.

3. **Frequency Distribution:** Bar plots can show the frequency or count of each category, helping to visualize the distribution of data.

4. **Presentation:** They are commonly used in presentations, reports, and publications because they provide a clear and straightforward way to convey information.

```python
import numpy as np
plt.figure(figsize=(7,5))
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit, color='#F429F9') 
plt.title('Company vs Profit(Barplot)')
plt.grid()
```

```python
import numpy as np
plt.figure(figsize=(10,6))
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit, color='#7729F9') 
plt.title('Company vs Profit (Horizontal Barplot)')
plt.grid()
```

Ans.5 A box plot, also known as a box-and-whisker plot, is a data visualization that provides a graphical summary of the distribution and key statistics of a dataset. It is used to display the spread, central tendency, and potential outliers within the data. Box plots are particularly useful for the following purposes:

1. **Summary of Data Distribution:** Box plots display important statistics such as the median, quartiles, and potential outliers, giving a concise summary of the data's distribution.

2. **Outlier Detection:** They help identify outliers, which are data points that are significantly different from the rest of the dataset.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)

# Add labels and a title
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot Example')

# Display the plot
plt.show()
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)

# Add labels and a title
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot Example')

# Display the plot
plt.show()
```
